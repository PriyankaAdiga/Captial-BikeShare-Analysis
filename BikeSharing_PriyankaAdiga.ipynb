###Predicting the trend of renting on hourly basis

```
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```
hour_dataframe.drop('instant',axis=1,inplace=True)
hour_dataframe.head()
```

```
merge_dataframe = pd.merge(day_dataframe,hour_dataframe,how='left',left_on='dteday',right_on='dteday')
merge_dataframe.drop('instant',axis=1, inplace=True)
merge_dataframe.head()
```

```
bike_X = merge_dataframe.drop(['dteday','cnt_y','casual_y','registered_y'], axis=1)
bike_Y = merge_dataframe['cnt_y']
```

```
merge_dataframe.columns
```

```
bike_Xtrain, bike_Xtest, bike_Ytrain, bike_Ytest = train_test_split(bike_X,bike_Y)
```

```
bike_Xtrain.columns
```

```
bike_model1 = LinearRegression()
```

```
bike_model1.fit(bike_Xtrain,bike_Ytrain)
```

```
bike_model_predict = bike_model1.predict(bike_Xtest)
bike_model_predict
```

```
plt.scatter(bike_Ytest,bike_model_predict)
plt.xlabel("Test Set")
plt.ylabel("Prediction")
plt.title("Bike Sharing Prediction scatter plot")
```

```
print('Mean Absolute Error-MAE= ', metrics.mean_absolute_error(bike_Ytest,bike_model_predict))
print('Mean Squared Error-MSE= ', metrics.mean_squared_error(bike_Ytest,bike_model_predict))
print('Root Mean Squared Error-RMS= ', np.sqrt(metrics.mean_squared_error(bike_Ytest,bike_model_predict)))
```

```
pd.DataFrame(data=bike_model1.coef_, index=bike_X.columns, columns=['Coefficient'])
```
